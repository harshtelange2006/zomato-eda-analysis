```python
# Average cost for two people per location
```

```python
# Clean cost column
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '').str.strip()
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')

# Average cost per location
df9 = df.groupby('location')['approx_cost(for two people)'].mean().sort_values(ascending=False).reset_index()
df9.columns = ['location', 'avg_cost']
df9.head(10)
```

```python
df9.head(10).plot(kind='bar', x='location', y='avg_cost', figsize=(10,5), color='orange')
plt.title("Top 10 Most Expensive Locations (Avg Cost for Two)")
plt.xlabel('Location')
plt.ylabel('Avg Cost (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
